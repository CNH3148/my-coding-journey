```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from ipywidgets import interact
```

```python
def draw(n):
    x = np.linspace(0, 2*np.pi, 1000)
    y = np.sin(n*x)
    plt.plot(x, y)
    plt.show()  # 一定要有這行，不然圖不會隨之改變
```

```python
draw(3)
```

```python
interact(draw, n=(1., 50.))
```
