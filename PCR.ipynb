```python
import matplotlib.pyplot as plt

# 定義時間和溫度數據
time = [0, 5, 6, 6, 6.5833, 6.5833, 10.0833, 17.0833, 17.0833, 27.0833]  # 時間數據（分鐘）
temperature = [94, 94, 94, 52, 52, 72, 72, 72, 16, 16]  # 溫度數據（攝氏度）

# 設定圖片大小
plt.figure(figsize=(10, 6))

# 繪製折線圖
plt.plot(time, temperature, marker='o', linestyle='-')

# 添加標籤和標題
plt.xlabel('Time (minutes)')
plt.ylabel('Temperature (°C)')
plt.title('PCR Time and Temperature Profile')

# 在每個階段的線段上方添加溫度和時長的標籤
for i in range(len(time) - 1):
    segment_time = time[i + 1] - time[i]
    if segment_time == 0.0:  # 檢查時長是否為 0
        continue  # 如果時長為 0，則跳過這個階段

    segment_temperature = temperature[i]
    label = f"{segment_temperature:.1f}°C, {segment_time:.1f} min"  # 格式化為小數點後一位

    # 調整標籤的位置以避免重疊
    va = 'bottom'
    if i % 2 == 1:  # 奇數階段向上調整位置
        va = 'top'
    plt.text((time[i] + time[i + 1]) / 2, temperature[i], label, ha='center', va=va)

# 顯示圖表
plt.grid()
plt.show()
```
