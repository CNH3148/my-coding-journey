# Array 導向的程式寫法

````python
# 一般用迴圈處理 List 的方法，
# 如果項目數很多，用迴圈處理資料的方法便不切實際。
USD = [1096.95, 596.95, 896.95]
exchange_rate = 30.1
for i in USD:
    print(i*exchange_rate)
````

````python
result = []
for i in USD:
    TWD = i*exchange_rate
    result.append(TWD)

result
````

## 數據分析最基本的理念之一：**能不要用迴圈就不要用 (∵計算速度)**

````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
````

````python
# 範例一
USD = [1096.95, 596.95, 896.95]
exchange_rate = 30.1
````

````python
prices = np.array(USD)
prices
````

````python
print(prices)
````

### array 可以做類似向量的運算！
不到最後關頭，不輕言使用 List！

````python
twd = prices * exchange_rate
twd
````

## Array 將資料轉成矩陣儲存

````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
````

````python
A = np.random.randn(100) # 標準常態分佈中抽取 100 個數；平均數 0，標準差 1
A
````

````python
# 改為平均 50，標準差 10
A = A * 10 + 50
A
````

````python
# 原本的型態：
A.shape
````

````python
# 改成 5 * 20 的矩陣
A.shape = (5, 20)
A
````

````python
A.reshape(100, 1)
````

````python
# .reshape() 不會改變原本的資料
A
````

````python
A = A.reshape(100, )
A
````

````python
A = A.reshape(20, 5)
A
````
